### Step 1. Importieren der notwendigen Pakete

```python
import pandas as pd
```

### Step 2. Importiere die Daten von folgender [Adresse](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Weise dem Datensatz die Variable chipo zu.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep="\t")
```

### Step 4. Wie viele Produkte kosten mehr als $10.00?

```python
chipo["item_price"] = chipo.item_price.str.slice(start=1).astype(float)
```

```python
chipo.query("item_price > 10")
```

### Step 5. Was ist der Preis jedes einzelnen Produkts? 
###### Gib einen Dataframe aus, der nur 2 Spalten enthält: item_name und item_price.

```python
chipo[["item_price", "item_name"]]
```

### Step 6. Sortiere den Dataframe anhand des item_name.

```python
chipo.sort_values(["item_price", "order_id"])
```

### Step 7. Wie oft wurde das teuerste Produkt bestellt?

```python
name = chipo.sort_values("item_price", ascending=False)["item_name"][0]
chipo.query("item_name == @name")
```

```python
max_price = chipo["item_price"].max()
chipo[chipo["item_price"] == max_price]["quantity"].sum()
```

### Step 8. Wie oft wurde ein (oder mehrere) Veggie Salad Bowl bestellt?

```python
len(chipo[chipo.item_name == "Veggie Salad Bowl"])
```

### Step 9. Wie oft haben Leute mehr als eine Canned Soda bestellt?

```python
chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
chipo.query("item_name == 'Canned Soda' and quantity > 1")
```
